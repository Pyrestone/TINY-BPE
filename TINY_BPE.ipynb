```python
text=" ".join(nltk.corpus.brown.words())
```

```python
text[:200]
```

```python
from tinybpe import tiny_bpe
```

```python
#Create an empty tokenizer
tokenizer=tiny_bpe()
```

```python
# reduce text to standard character set
text_normalized=tokenizer.normalize_text(text)
text_normalized[:200]
```

```python
#Train tokenizer on the text.
#Runtime is O(input_len * merge_count)
input_len=100000
merge_count=1000


tokenizer.fit_vocab(text_normalized[:input_len],merge_count,verbose=1) #set verbose=0 to not print output
```

```python
# Save tokenizer to file

with open("tokenizer_brown_100000_1000.json","w") as file:
    json.dump(tokenizer.serialize(),file)
```

```python
del tokenizer
```

```python
# Load trained tokenizer from file

tokenizer=tiny_bpe()

with open("tokenizer_brown_100000_1000.json","r") as file:
    tokenizer.deserialize(json.load(file))
```

```python
test_text="Contrary to popular belief, Lorem Ipsum is not simply random text. It has roots in a piece of classical Latin literature from 45 BC, making it over 2000 years old."
print(test_text)
test_text=tokenizer.normalize_text(test_text)
print(test_text)
```

```python
#you can tokenize a string into a list of tokens
print("|".join(tokenizer.string_tokenize(test_text)))
```

```python
#you can also get an array of integers, which can be used for word embeddings.
print(tokenizer.tokenize(test_text))
```

```python
# The merge probability can be used to reduce overfitting on whole words 
# by not always performing the merges

# The below example shows the effects of different merge probabilities.
# Note that merge_probs other than 0 or 1 produce non-deterministic output.

for i in range(20):
    print("merge probability:",(i+1)/20)
    tokenized_string=tokenizer.string_tokenize(test_text,merge_prob=(i+1)/20) # List of token strings
    tokenized_arr=tokenizer.tokenize(test_text,merge_prob=((i+1)/20))      # Int array
    print("|".join(tokenized_string))
    print()
```
